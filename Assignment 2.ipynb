```python


mean_age = 38
std_dev_age = 6

prob_older_than_44 = 1 - norm.cdf(44, loc=mean_age, scale=std_dev_age)
prob_between_38_and_44 = norm.cdf(44, loc=mean_age, scale=std_dev_age) - norm.cdf(38, loc=mean_age, scale=std_dev_age)

print("Probability of being older than 44:",round(prob_older_than_44*100,2),'%')
print("Probability of being between 38 and 44:",round(prob_between_38_and_44*100,2),'%')

print("More employees are older than 44 than between 38 and 44: {prob_older_than_44 > prob_between_38_and_44}")

prob_under_30 = norm.cdf(30, loc=mean_age, scale=std_dev_age)
expected_number_under_30 = prob_under_30 * 400

print("Expected number of employees under the age of 30: {expected_number_under_30:.2f}")
```

```python
mew = 50
std = 40
n = 100
lower_limit = 45
upper_limit = 55

# Calculate z-scores
z_score_lower = (lower_limit - mew) / (std / (n**0.5))
z_score_upper = (upper_limit - mew) / (std / (n**0.5))

# Calculate probabilities
probability_investigation = stats.norm.cdf(z_score_lower) + (1 - stats.norm.cdf(z_score_upper))

print(round(probability_investigation*100,2),"%")
```
